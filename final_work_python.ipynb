```python
limit = 2 # устанавливаем лимит для границ (откуда рисуем)
step = 0.01 
color = 'b'
line_s = '-'
direct_up = True # маркер направления (up - down)
```

```python
a, b, c, d, e = -12, -18, 5, 10, -30
x = np.arange(-limit, limit, step)
```

```python
def switch_color():
    global color
    if color == 'b':
        color = 'r'
    else:
        color = 'b'
    return color
```

```python
def switch_line(): # f > 0 или f < 0
    global line_s
    if line_s == '-':
        line_s = '--'
    else:
        line_s = '-'
    return line_s
```

```python
def func(x):
    return a * x**4 + b *x**3 + c *x **2 + d * x + e
```

```python
plt.rcParams['lines.linestyle'] = '-'
x_lim = np.arange(-1000, 1000, step)
plt.plot(x_lim, func(x_lim), color)
plt.grid()
```
